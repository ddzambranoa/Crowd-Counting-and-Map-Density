```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.title("Gáfico de prueba")
```

```python
y = np.linspace(0, 10, 100)
x = np.arange(len(y))
y1 = np.sin(y)

plt.figure(figsize=(10,5))
plt.title("Gáfico de prueba")
plt.plot(x, y, color="#6cb415", label="linear")
plt.scatter(x, y1, s=y, color="black", label="sin")
plt.grid()
plt.legend(loc=5)
plt.xlabel("Eje X") # con esta línea generamos el texto en el eje x
plt.ylabel("Eje Y");
```

```python
plt.figure(figsize=(10,5), layout="constrained")
plt.subplot(2, 2, 1)
plt.plot(x, y, color="#6cb415", label="linear")
plt.grid()
plt.title("Gráfico 1")

plt.subplot(2, 2, 2)
plt.scatter(x, y1, s=y, color="black", label="sin")
plt.xlabel("Eje X")
plt.title("Gráfico 2", fontsize=8)

plt.subplot(2, 2, 3)
plt.plot(x, x**3, color="blue", label="linear")
plt.grid()
plt.title("Gráfico 3")

plt.subplot(2, 2, 4)
plt.plot(x, y, color="red", label="linear")
plt.grid()
plt.title("Gráfico 3")
```

```python
ncols = 2
nrows = 2
w = 5*ncols
h = 3*nrows
plt.figure(figsize=(w,h), layout="constrained")
for fig in range(ncols*nrows):
    plt.subplot(nrows, ncols, fig+1)
    plt.plot(x, y**(fig), color="#6cb415", label="linear")
    plt.grid()
    plt.title(f"Gráfico {fig+1}")
```

```python
plt.scatter(x, y1, s=y, color="red", label="sin")
plt.xlabel("Eje X")
plt.title("Gráfico 2", fontsize=8)

plt.savefig("./sin.png")
plt.show()
```

```python
plt.plot(x, x, color="black", label="label 1", ls=":")
plt.plot(x, x*2, color="black", label="label 2", ls="--")
plt.plot(x, x*3, color="black", label="label 3", ls="-.")
plt.plot(x, x*4, color="black", label="label 4", ls="-")
plt.xlabel("Eje X")
plt.legend()
plt.title("Gráfico 2", fontsize=8)
```

```python
plt.bar(x[:3], x[:3]+4, color="red", label="sin")
plt.xlabel("Eje X")
plt.xticks(x[:3], ["CARROS", "B", "C"])
plt.title("Gráfico 2", fontsize=8)
#
```
